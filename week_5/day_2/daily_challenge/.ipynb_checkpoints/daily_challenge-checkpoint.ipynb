```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
starbucks = pd.read_csv('starbucks.csv')
```

```python
starbucks
```

```python
starbucks.head(10)
```

```python
print(f'Total number of entries in starbucks: {starbucks.shape[0]}')
```

```python
print(starbucks.columns)
```

```python
starbucks.index
```

```python
drink_with_most_sugar = starbucks.groupby('Beverage')[' Sugars (g)'].mean().idxmax()
```

```python
print(f'The drink with the most sugar on average: {drink_with_most_sugar}')
```

```python
print("Number of different drink categories:", starbucks['Beverage_category'].nunique())
```

```python
print('Top 5 most frequent drink categories:', starbucks['Beverage_category'].value_counts().head(5))
```

```python
print('Average calorie content per drink category:', starbucks.groupby('Beverage_category')['Calories'].mean())
```

```python
starbucks['Caffeine (mg)'] = pd.to_numeric(starbucks['Caffeine (mg)'], errors='coerce').round().astype('Int64')
```

```python
starbucks['Caffeine (mg)']
```

```python
starbucks['caffeine_to_calories'] = (starbucks['Caffeine (mg)'] / starbucks['Calories']).round().astype('Int64')
```

```python
starbucks['caffeine_to_calories']
```

```python
print('Drink category with the highest caffeine-to-calories ratio:', starbucks.groupby('Beverage_category')['caffeine_to_calories'].mean().idxmax())
```

```python
plt.figure(figsize=(10, 6))
starbucks.groupby('Beverage_category')['caffeine_to_calories'].mean().sort_values().plot(kind='barh')
plt.xlabel('Average Caffeine to Calories Ratio')
plt.ylabel('Drink Category')
plt.title('Average Caffeine to Calories Ratio for Each Drink Category')
plt.show()
```
