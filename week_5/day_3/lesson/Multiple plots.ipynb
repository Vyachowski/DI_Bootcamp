```python
x = np.linspace(1, 100, 1000)
y1 = x ** 2
y2 = x ** 3

figure, axes_set = plt.subplots(2)

axes_set[0].plot(x, y1, color='red')

axes_set[0].annotate('breakout!', xy=(0,0), xytext=(1,2000), 
                    arrowprops={'facecolor':'black', 'shrink':10})

axes_set[1].plot(x, y2, color='green')

plt.show()
```

```python
x = np.linspace(1, 100, 1000)
y1 = x ** 2
y2 = x ** 3

figure, axes_set = plt.subplots(2, sharex=True)

axes_set[0].plot(x, y1, color='red', label='main 1 line')
axes_set[0].legend(title='Legend', loc='lower right', shadow=True)

axes_set[0].annotate('breakout!', xy=(0,0), xytext=(1,2000), 
                    arrowprops={'facecolor':'black', 'shrink':10})

axes_set[1].plot(x, y2, color='green', label='main 2 line')
axes_set[1].legend()

plt.tight_layout()
```

7. Create a figure with four subplots. In the first subplot, plot y against x. In the second subplot, plot -y against x. In the third subplot, plot y squared against x. In the fourth subplot, plot -y squared against x.

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
figure, axes_array = plt.subplots(4)
axes_array[0].plot(x, y)
axes_array[1].plot(x, -y)
axes_array[2].plot(x, y**2)
axes_array[3].plot(x, -y**2)

plt.tight_layout()
```

9. For each subplot, add a legend that explains what each line represents. You can set the legend title as Legend and place it in the upper right corner of each subplot. Enable the shadow effect for the legend.

```python
figure, axes_array = plt.subplots(4)
axes_array[0].plot(x, y, label='X,Y')
axes_array[0].legend(title='Legend', loc='upper right')
axes_array[1].plot(x, -y, label='X, -Y')
axes_array[1].legend()
axes_array[2].plot(x, y**2, label='X, Y^2')
axes_array[2].legend()
axes_array[3].plot(x, -y**2, label='X, -(Y^2)')
axes_array[3].legend()


axes_array[0].annotate(np.max(y), xy=(x[max_y_idx],y[max_y_idx]), xytext=(x[max_y_idx]+1,y[max_y_idx]+0.5), 
                    arrowprops={'facecolor':'black', 'shrink':10})
axes_array[0].annotate(np.min(y), xy=(x[min_y_idx],y[min_y_idx]), xytext=(x[min_y_idx]-1,y[min_y_idx]-0.5), 
                    arrowprops={'facecolor':'black', 'shrink':10})

plt.tight_layout()
```
