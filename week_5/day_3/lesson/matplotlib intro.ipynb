```python
# Create a line plot for the function y = x^2 for x ranging from 0 to 10.

x = np.linspace(0, 10, 100)
y = x ** 2

figure, axes = plt.subplots()
axes.plot(x, y)
```

```python
# 2. Create a line plot of y1 against x.
figure, axes = plt.subplots()
axes.plot(x, y1)
```

```python
# 2. Create a scatter plot of y2 against x.
figure, axes = plt.subplots()
axes.scatter(x, y2)
```

```python
# 3. Save the plot you created in Exercise 2 as a PNG file named my_plot.png.
figure.savefig('my_plot.png')
```

```python
# 4. - Use the same data as in Exercise 2.
x = np.linspace(0, 10, 20)
y1 = x ** 2
y2 = x ** 1.5
```

```python
# - Create a line plot of y1 against x with a red dashed line and circular markers. Label it y = x^2.
figure, axes = plt.subplots()
axes.set(title='y = x^2')
axes.plot(x, y1, marker='o', linestyle='dashed')
```

```python
# - Create a line plot of y2 against x with a blue dotted line and square markers. Label it y = x^1.5.
figure, axes = plt.subplots()
axes.set(title='y = x^1.5')
axes.plot(x, y2, marker='s', color='blue', linestyle='dotted', label='y = x^1.5')
```

```python
# - Set the title as My Styled Plot and label the axes as x and y.
figure, axes = plt.subplots()
axes.set(title='My Styled Plot', xlabel='X', ylabel='Y')
axes.plot(x, y2, marker='s', color='blue', linestyle='dotted')
```

```python
# - Add a legend to the plot.
# - Set the title as My Styled Plot and label the axes as x and y.
figure, axes = plt.subplots()
axes.set(title='My Styled Plot', xlabel='X', ylabel='Y')
axes.plot(x, y2, marker='s', color='blue', linestyle='dotted', label='Blue line')
axes.legend()
```

```python
# - Save the plot as my_styled_plot.png.
figure.savefig('my_styled_plot.png')
```
