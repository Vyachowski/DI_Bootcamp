# Exercise 8
Instructions
Plot the line plot and scatter plot from Exercise 7 on the same figure using subplots. Use plt.subplots to create a figure with two subplots.

```python
values = np.linspace(-2, 2, 100)

def f(x):
    return x**4

y_values = f(values)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

line_plot = ax1.plot(values, y_values, color='tab:blue')

scatter_plot = ax2.scatter(values, y_values, color='tab:blue', marker='p')

plt.show()
```

# Exercise 9
Instructions
Modify the plots from Exercise 8 to include a legend. Use the legend function from Matplotlib.

```python
values = np.linspace(-2, 2, 100)

def f(x):
    return x**4

y_values = f(values)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

line_plot = ax1.plot(values, y_values, color='tab:blue', label='$f(x) = x^4$')
ax1.legend()

scatter_plot = ax2.scatter(values, y_values, color='tab:blue', marker='p', label='$f(x) = x^4$')
ax2.legend(loc='lower right')

plt.show()
```
