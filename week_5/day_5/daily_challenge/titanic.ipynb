```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

## Female to Male Ratio

```python
ratio = df['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(ratio, labels=ratio.index, autopct='%1.1f%%', startangle=90)
plt.title('Male to female Ratio')
plt.show()
```

```python
## Passengers age
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'].dropna(), bins=20, edgecolor='black')
plt.xlabel('Age')
plt.title('Distribution of passenger ages')
plt.show()
```

```python
## Survived bar
```

```python
survival_percentage = df['Survived'].value_counts(normalize=True) * 100
plt.figure(figsize=(6, 6))
plt.bar(survival_percentage.index, survival_percentage.values, color=['black', 'red'])
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.ylabel('Percentage')
plt.title('Survival Percentage')
plt.ylim(0, 65)
plt.show()
```
