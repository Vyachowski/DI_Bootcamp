```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

## Female to Male Ratio

```python
ratio = df['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(ratio, labels=ratio.index, autopct='%1.1f%%', startangle=90)
plt.title('Male to female Ratio')
plt.show()
```

```python
## Passengers age
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'].dropna(), bins=20, edgecolor='black')
plt.xlabel('Age')
plt.title('Distribution of passenger ages')
plt.show()
```

```python
## Survived bar
```

```python
survival_percentage = df['Survived'].value_counts(normalize=True) * 100
plt.figure(figsize=(6, 6))
plt.bar(survival_percentage.index, survival_percentage.values, color=['black', 'red'])
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.ylabel('Percentage')
plt.title('Survival Percentage')
plt.ylim(0, 65)
plt.show()
```

```python
## Survived bar by gender
```

```python
survival_gender = df.groupby(['Survived', 'Sex']).size().unstack()
survival_gender.plot(kind='bar', stacked=True, color=['grey', 'black'], figsize=(8, 6))
plt.xticks([0, 1], ['Not survived', 'Survived'])
plt.ylabel('Number of Passengers')
plt.title('Survival by gender')
plt.show()
```

```python
## Scatter plot fare vs age
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Correlation between Fare and Age')
plt.show()
```

## Normalizing data. Missing values. MinMaxScaler. Encoding.

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
scaler = MinMaxScaler()
df['Fare'] = scaler.fit_transform(df[['Fare']])
encoder = OneHotEncoder(drop='first', sparse_output=False)
sex_encoded = encoder.fit_transform(df[['Sex']])
sex_encoded_df = pd.DataFrame(sex_encoded, columns=['Male'])
df = pd.concat([df, sex_encoded_df], axis=1)
df.drop(columns=['Sex'], inplace=True)
```

## Data Analysis

### Survival Rate

```python
survival_rate = df.groupby('Male')['Survived'].mean() * 100
```

```python
print("Survival Rate:", survival_rate)
```

```python
class_survival_rate = df.groupby('Pclass')['Survived'].mean() * 100
```

```python
print("Survival Rate by class:", class_survival_rate)
```
